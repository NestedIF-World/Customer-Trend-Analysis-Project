```python
#To import necessary libraries
import pandas as pd
```

```python
#To read the file
df = pd.read_csv(r"C:\Users\ritik\Downloads\customer_shopping_behavior.csv")
```

```python
#To know how our data looks like
df.head()
```

```python
#To know the datatypes
df.dtypes
```

```python
#Statistical Analysis
df.describe().T 
```

```python
#To check the null values
df.isnull().sum()
```

```python
df['Review Rating']=df['Review Rating'].fillna(df.groupby('Category')['Review Rating'].transform('median'))
```

```python
#To check if any null values present
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
#To create 'age_group' column
labels=['young','middle-aged','adult','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
#To check if age_group is created or not
df[['age','age_group']]
```

```python
#To check how many unique values are present
df['frequency_of_purchases'].unique()
```

```python
#To create column purchase frequency days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 months':90
}
df['purchase_frequency']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
#To check if column is created
df[['purchase_frequency','frequency_of_purchases']]
```

```python
df[['discount_applied','promo_code_used']].head()
```

```python
#To check if both these columns carry same values
df['discount_applied'].equals(df['promo_code_used'])
```

```python
#Since both the columns carry same values, we'll drop the column 'promo_case_used'
df.drop('promo_code_used',axis=1,inplace=True)
```
